```python
df_cov_glue = pd.read_csv("cov-glue-substitutions.csv").set_index("replacementID")
```

```python
df_nextclade = pd.read_csv("nextclade-fixed.csv")
```

```python
protein_to_counts_df = {
    prot.replace(" ", ""): sub_df
    for (prot, sub_df) in df_cov_glue.groupby("genomeRegion")
    
}
```

```python
from collections import Counter 
c = Counter() 
for _, row in df_nextclade.iterrows():
    print("\n---\n%s\n---" % row.seqName)
    for mut in row.mutations.split(","):
        print("%s" % mut)
        c[mut] += 1
```

```python
with open("counts-table.txt", "w") as f:
    f.write("%12s %12s %20s %20s %20s\n" % ("Protein", "Mutation", "Number in Cohort", "Number in CoV-GLUE"))
    for mut, count in c.most_common():
        prot, subst = mut.split(":")
        covglue = protein_to_counts_df.get(prot)
        if covglue is None:
            num_in_covglue = 0
        else:
            num_in_covglue_df = covglue[covglue.replacement == subst]
            num_in_covglue = num_in_covglue_df.numSeqs.sum()
        f.write("%12s %12s %20s %20s\n" % (prot, subst, count, num_in_covglue))
with open("counts-table.txt") as f:
    print(f.read())
```
