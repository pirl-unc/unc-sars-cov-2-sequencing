```python
orf1ab_nsp_to_seq = {}
last_id = None
buffer = []

def add_last_seq_and_clear_buffer():
    if last_id:
        orf1ab_nsp_to_seq[last_id] = "".join(buffer)
        buffer.clear()
    elif len(buffer) > 0:
        raise ValueError("Unexpected data in buffer")
        
with open("orf1ab-non-structural-proteins.fasta") as f:

    for l in f:
        if l.startswith(">"):
            add_last_seq_and_clear_buffer()
            last_id = l.split()[1]
        else:
            l = l.strip()
            if l:
                buffer.append(l)
    add_last_seq_and_clear_buffer()
orf1ab_nsp_to_len = {k: len(s) for k, s in orf1ab_nsp_to_seq.items()}
orf1ab_nsp_to_len
```

```python
ORF1ab_len = len(ORF1ab_seq)
```

```python
base0_coord_to_nsp_name = [None] * ORF1ab_len
amino_acids_before_nsp = {}
last_pos = 0
for nsp_name in nsps_in_order:
    amino_acids_before_nsp[nsp_name] = last_pos
    n = orf1ab_nsp_to_len[nsp_name]
    for i in range(n):
        base0_coord_to_nsp_name[i + last_pos] = nsp_name
    last_pos += n
```

```python
# sanity check mapping from ORF1b to nsp sequences
for i in range(2700):
    orf1a_offset = amino_acids_before_nsp["nsp12"] + 9
    pos_in_orf1ab = i + 1 + orf1a_offset
    nsp_name = base0_coord_to_nsp_name[pos_in_orf1ab - 1]
    nsp_pos = pos_in_orf1ab - amino_acids_before_nsp[nsp_name]
    nsp_aa = orf1ab_nsp_to_seq[nsp_name][nsp_pos - 1]
    ORF1ab_aa = ORF1ab_seq[pos_in_orf1ab - 1]
    assert nsp_aa == ORF1ab_aa
    
    if nsp_name == "nsp16":
        print("ORF1b", i + 1, nsp_name, nsp_pos, nsp_aa, "ORF1ab", pos_in_orf1ab, ORF1ab_aa)
    
```

```python
import unidecode

def normalize_string(s, strict_ascii=False):
    if type(s) is float and np.isnan(s):
        return ""
    if strict_ascii:
        # strip out weird unicode spaces
        char_codes = [ord(c) for c in s]
        ascii_char_codes = [c for c in char_codes if c <= 128]
        chars = [chr(c) for c in ascii_char_codes]
        s_without_unicode = "".join(chars)
    else:
        s_without_unicode = unidecode.unidecode(s)

    s = s.strip()
    if s != s_without_unicode:
        print("Converted %s (%d chars) => %s (%d chars)" % (
            s, len(s),
            s_without_unicode, len(s_without_unicode)))
    return s_without_unicode

def normalize_list(s):
    s = normalize_string(s)
    parts = s.split(",")
    parts = [p.strip() for p in parts]
    return [p for p in parts if p]

def normalize_int(s):
    if type(s) is int:
        return s
    elif type(s) is float:
        return int(s)
    else:
        assert type(s) is str
        s_without_unicode = drop_invisible_chars(s)
        return int(s_without_unicode)

from collections import Counter
mut_counts = Counter()
rows = []
for row_num , old_row in df.iterrows():
    new_row = {}
    seqname = normalize_string(old_row.seqName.split(".")[0])
    new_row["seqName"] = seqname
    new_row["clade"]  = normalize_string(old_row.clade)
    new_row["substitutions"] = normalize_string(old_row.substitutions) 
    new_row["deletions"] = normalize_string(old_row.deletions) 
    new_row["insertions"] = normalize_string(old_row.insertions) 
    new_row["missing"] = normalize_string(old_row.missing) 
    new_row["pcrPrimerChanges"] = normalize_string(old_row.pcrPrimerChanges)
    new_row["totalPcrPrimerChanges"] = normalize_int(old_row.totalPcrPrimerChanges)
    new_row["totalMutations"] = normalize_int(old_row.totalMutations)
    dels = normalize_list(old_row["aaDeletions"])
    if dels:
        assert False, "Found deletions %s for %s" % (dels, old_row["seqName"])
    muts = normalize_list(old_row["aaSubstitutions"])
    print("=== %s ===" % seqname)
    new_muts = []
    for mut in muts:
        prot, subst = mut.split(":")
        ref = subst[0]
        alt = subst[-1]
        pos = normalize_int(subst[1:-1])
        
        if prot == "ORF14":
            print("\t Skipping %s" % (mut,))
            continue 
            
        if prot in {"ORF1a", "ORF1ab"}:
            nsp_name = base0_coord_to_nsp_name[pos - 1]
            nsp_pos = pos - amino_acids_before_nsp[nsp_name]
        elif prot == "ORF1b":
            nsp12_offset = amino_acids_before_nsp["nsp12"]
            pos_in_orf1ab = pos + nsp12_offset + 9
            
            nsp_name = base0_coord_to_nsp_name[pos_in_orf1ab - 1]
            nsp_pos = pos_in_orf1ab - amino_acids_before_nsp[nsp_name]
            nsp_aa = orf1ab_nsp_to_seq[nsp_name][nsp_pos - 1]
            ORF1ab_aa = ORF1ab_seq[pos_in_orf1ab - 1]    
            assert nsp_aa == ORF1ab_aa
        else:
            nsp_name = nsp_pos = None
        
        if nsp_name:
            print("\t Mapping %s => %s %d" % (mut, nsp_name, nsp_pos))
            old_mut = mut
            new_mut = "%s:%s%d%s" % (nsp_name, ref, nsp_pos, alt)
            prot = nsp_name
            pos = nsp_pos
        else:
            old_mut = ""
            new_mut = mut
        
        if "X" in new_mut:
            old_mut = new_mut
  
            new_mut = old_mut.replace("X", "fs")
            print("\t Renaming %s => %s" % (old_mut, new_mut))
        
        mut_counts[new_mut] += 1
        new_muts.append(new_mut)
    new_row["aaSubstitutionsOld"] = old_row["aaSubstitutions"]
    new_row["mutations"] = ",".join(new_muts)
    rows.append(new_row)

print("Processed %d records" % (len(rows),))

df_corrected = pd.DataFrame.from_records(rows)
```
